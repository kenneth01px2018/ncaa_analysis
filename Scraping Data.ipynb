```python
import pandas as pd
url = 'https://basketball.realgm.com/ncaa/team-stats/2022/Advanced_Stats/Team_Totals/0'
pd.read_html(url)[0]
advanced = pd.read_html(url)[0]
del advanced[advanced.columns[0]]
advanced
```

```python
url = 'https://basketball.realgm.com/ncaa/team-stats/2022/Averages/Team_Totals/0'
pd.read_html(url)[0]
averages = pd.read_html(url)[0]
del averages[averages.columns[0]]
averages
```

```python
url = 'https://basketball.realgm.com/ncaa/team-stats/2022/Misc_Stats/Team_Totals/0'
misc = pd.read_html(url)[0]
del misc[misc.columns[0]]

misc = misc.drop(columns = ['Dbl Dbl', 'Tpl Dbl', '40 Pts', '20 Reb', '20 Ast', 
                                         '5 Stl', '5 Blk', 'High Game', 'MOV', 'Techs'])
misc
```

```python
temp = pd.merge(averages, advanced, how = 'left', on = ['Team'])
combined = pd.merge(temp, misc, how = 'left', on = ['Team'])
combined
```

```python
combined.nunique(axis = 0)
```

```python
url = 'https://basketball.realgm.com/ncaa/teams/2022'
conf = pd.read_html(url)[0]
conf
```

```python
combined.to_csv('./Data/ncaa22.csv')
conf.to_csv('./Data/conf22.csv')
```
